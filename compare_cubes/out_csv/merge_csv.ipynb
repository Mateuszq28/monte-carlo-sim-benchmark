```python
print_stats_srodkowe_przekroje_2D = [f for f in fl if 'print_stats_srodkowe_przekroje_2D' in f]
print_stats_srednie_2D = [f for f in fl if 'print_stats_srednie_2D' in f]
print_stats = [f for f in fl if ('print_stats_' in f and f not in print_stats_srednie_2D and f not in print_stats_srodkowe_przekroje_2D)]






compare_all_cubes_universal_srodkowe_przekroje_2D = [f for f in fl if 'compare_all_cubes_universal_srodkowe_przekroje_2D' in f]
compare_all_cubes_universal_srednie_2D = [f for f in fl if 'compare_all_cubes_universal_srednie_2D' in f]
compare_all_cubes_universal = [f for f in fl if ('compare_all_cubes_universal_' in f and f not in compare_all_cubes_universal_srednie_2D and f not in compare_all_cubes_universal_srodkowe_przekroje_2D)]
```

```python
sublists = [print_stats_srodkowe_przekroje_2D,
            print_stats_srednie_2D,
            print_stats,
            compare_all_cubes_universal_srodkowe_przekroje_2D,
            compare_all_cubes_universal_srednie_2D,
            compare_all_cubes_universal]
```

```python
lens = [len(s) for s in sublists]
lens
```

```python
del_prefix = ['print_stats_srodkowe_przekroje_2D',
              'print_stats_srednie_2D',
              'print_stats_',
              'compare_all_cubes_universal_srodkowe_przekroje_2D',
              'compare_all_cubes_universal_srednie_2D',
              'compare_all_cubes_universal_']
```

```python
sublists1 = sublists[:3]
del_prefix1 = del_prefix[:3]

sublists2 = sublists[3:]
del_prefix2 = del_prefix[3:]
```

```python
df1 = pd.DataFrame()
for s, d in zip(sublists1, del_prefix1):
    for filename in s:
        new = pd.read_csv(filename)
        new['table_category'] = d.strip('_ ')
        new['experiment_seria'] = filename[len(d):-4].strip('_ ')
        df1 = pd.concat([df1, new])

df2 = pd.DataFrame()
for s, d in zip(sublists2, del_prefix2):
    for filename in s:
        new = pd.read_csv(filename)
        new['table_category'] = d.strip('_ ')
        new['experiment_seria'] = filename[len(d):-4].strip('_ ')
        df2 = pd.concat([df2, new])
```

```python
df1.to_csv('all_print_stats.csv')
df2.to_csv('all_compare2benchmark.csv')
```
