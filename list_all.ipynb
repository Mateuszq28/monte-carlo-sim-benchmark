```python
import os
import pandas as pd
```

```python
fl = []
def list_files_scandir(path='.'):
    with os.scandir(path) as entries:
        for entry in entries:
            if entry.is_file():
                print(entry.path)
                fl.append(entry.path)
            elif entry.is_dir():
                add_only_dir = ['./compare_cubes\\models']
                if entry.path in add_only_dir:
                    fl.append(entry.path)
                elif entry.path not in ['./.git', './compare_cubes\\CUBES - Copy - all', './compare_cubes\\CUBES', './compare_cubes\\.ipynb_checkpoints', './compare_cubes\\__pycache__']:
                    list_files_scandir(entry.path)
                    fl.append(entry.path)

# Specify the directory path you want to start from
directory_path = './'
list_files_scandir(directory_path)
print(len(fl))
```

```python
print('len(fl)', len(fl))



path_lists = [path[2:] for path in fl]
print(len(path_lists))
remove_list = ['compare_cubes\\CUBES.zip']
path_lists = [e for e in path_lists if e not in remove_list]
print(len(path_lists))
path_lists = [path.split('\\') for path in path_lists]
path_lists = [[""]+path if len(path)==1 else path for path in path_lists]
path_lists = [[path[0], ' / '.join(path[1:])] for path in path_lists]


print('len(path_lists)', len(path_lists))
```
